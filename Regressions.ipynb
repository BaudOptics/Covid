# Covid Risk Factors Regression Analysis

```python
# Import dependencies

import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
# Import dataset

# You can pass a url to pd.read_csv, meaning we don't have to keep a copy of the csv downloaded

# df = pd.read_csv('http://data.cdc.gov/api/views/vbim-akqf/rows.csv?accessType=DOWNLOAD')

# This will take a LONG time - but it offers the advantage 
# that ANYONE can run this code without uploading a 1Gb csv to github

df = pd.read_csv('/home/harry/Projects/Covid/Covid.csv')

#
    
df
```

```python
#Ease of visibility
renames = {'Black, Non-Hispanic': 'Black', 'White, Non-Hispanic': 'White', 'Hispanic/Latino': 'Hispanic',
           'Asian, Non-Hispanic': 'Asian', 'Multiple/Other, Non-Hispanic':'Other', 
           'American Indian/Alaska Native, Non-Hispanic': 'Native', 
           'Native Hawaiian/Other Pacific Islander, Non-Hispanic': 'Pacific Islander'}
for key in renames.keys():
    df = df.replace(key, renames[key])
df
```

## Basic Data Cleanup

```python
#Because we will apply regression analysis to this data, we need represents our variables of interest as numbers

# Meaningless column
try:
    del df['Unnamed: 0']
except:
    df = df
    
df = df.replace(['Missing', 'Unknown'], np.nan) #no interest in these rows
df = df.replace('No', 0)
df = df.replace('Yes', 1)

df = df.replace('Male', 1)
df = df.replace('Female', 0)

df['sex'] = df['sex'].replace('Other', np.nan) #Negligle - only like 30 people are "other" in this dataset

#We don't mind keeping nan values for date data
df['pos_spec_dt'] = df['pos_spec_dt'].replace(np.nan, '--')
df['onset_dt'] = df['onset_dt'].replace(np.nan, '--')
df = df.dropna()
```

```python
#Still half a million rows of data with every desired datapoint available
df.columns
```

## Dummy Variable Classification
Dummy variables are a great way to let the regression software do the heavy lifting in the analysis of causal factors 

The underlying math is actually the same as a groupby average, but it gives our results in terms of regression coefficients, so we can very easily use them to estimate the future

https://conjointly.com/kb/dummy-variables/

```python
# pandas.get_dummies is custom-built for this

# Age:
ageDummies = pd.get_dummies(df['age_group'])
df = pd.merge(df, ageDummies, left_index=True, right_index=True)

# Ethnicity:
ethnoDummies = pd.get_dummies(df['Race and ethnicity (combined)'])
df = pd.merge(df, ethnoDummies, left_index=True, right_index= True)

#Ease of visibility

df
```

```python
#REGRESSION: LINEAR PROBABILITY MODEL

# Define input variables
x = df[['sex', 'medcond_yn', '0 - 9 Years', '10 - 19 Years', '20 - 29 Years',
       '30 - 39 Years', '40 - 49 Years', '50 - 59 Years', '60 - 69 Years',
       '70 - 79 Years', '80+ Years', 'Native', 'Asian','Black', 'Hispanic',
       'Other','Pacific Islander','White']]

# Avoid perfect multicollinearity (comment out these two lines and you'll see that it breaks the regression)
del x['0 - 9 Years']
del x['White']

# Define output variable (death)
y = df['death_yn']

# Ordinary Least Squares Regression
model = sm.OLS(y, x, hasconst= False)
res = model.fit()
res.summary()
```

## Regressing Death: Results
    Takeaways:
        - Men are 2.45% more likely to die than women (all else equal)
        - Those with a medical condition are 3.68% more likely to die than those without (all else equal)
        - Age's effect on the likelihood of death is shown in fig. 1
        - Ethnicity's effect is shown in fig. 2
        

```python
#Figure 1 calculation
age_coefs = {
'10 - 19 Years' :-0.0372 ,
'20 - 29 Years' :-0.0367 ,
'30 - 39 Years' :-0.0370 ,
'40 - 49 Years' :-0.0303 ,
'50 - 59 Years' :-0.0064 ,
'60 - 69 Years' :0.0525 ,
'70 - 79 Years' :0.1562,
'80+ Years'   :0.3556
}
master = {'Age': [], 'Coefficient': []}
for key in age_coefs.keys():
    master['Age'].append(key)
    master['Coefficient'].append(age_coefs[key])

ageDF = pd.DataFrame(master)
plt.figure(figsize= (20,8))
plt.plot(ageDF['Age'], ageDF['Coefficient'], marker ='o')
plt.xlabel('Age Group')
plt.ylabel('Coefficient')
plt.title('Fig. 1: Age\'s effect on risk of death')
plt.show()
```

```python
# Figure 2 calculations
races = {
'Native' :0.0450,
'Asian':0.0528 ,
'Black' :0.0510 ,
'Hispanic' :0.0286,
'Other' :0.0207 ,
'Pacific Islander' :0.0158 
}
raceDF = {'Race / Ethnicity': [], 'Coefficient':[]}


for key in races.keys():
    raceDF['Race / Ethnicity'].append(key)
    raceDF['Coefficient'].append(races[key])
    
plt.figure(figsize= (20,8))
x = raceDF['Race / Ethnicity']
y = raceDF['Coefficient']

plt.bar(x, y)
plt.ylabel('Coefficients')
plt.title('Fig. 2: Race\'s effect on risk of death')

plt.show()
```

```python
# % of patients who died by ethnicity demographic
total = pd.DataFrame(df.groupby('Race and ethnicity (combined)').sum()['death_yn']).rename(columns={'death_yn':'Total Deaths'})
counts = pd.DataFrame(df.groupby('Race and ethnicity (combined)').count()['death_yn']).rename(columns={'death_yn':'Total Observations'})
 
race = pd.merge(total, counts, left_index= True, right_index= True)
race['Pct Risk'] = (race['Total Deaths'] / race['Total Observations']) * 100
race['Pct Risk'].plot(kind='bar', ylabel='Pct Risk', figsize= (20,8))
plt.xticks(rotation='horizontal')
plt.show()
```

# Over/Under Representation
What groups are overrrepresented and underrepresented in mortality?

(these visualizations are a little bit ugly because I just made a function due to laziness -- will fix later)

```python
def popAndDeath(group):
    '''This function takes a parameter to group by
        and contrasts the parameter's population share and death share
    '''
    fig, axs = plt.subplots(1, 2, figsize =(18,10))

    deaths = df.groupby(group).sum()['death_yn']
    axs[1].pie(x= deaths, labels = deaths.index, textprops={'fontsize': 12} )
    axs[1].set_title('Share of Death:')

    counts = df.groupby(group).count()['death_yn']
    axs[0].pie(x= counts, labels = counts.index, textprops={'fontsize': 12})
    axs[0].set_title('Share of Population')

    plt.show()
```

```python
popAndDeath('Race and ethnicity (combined)')
```

```python
popAndDeath('age_group')
```

```python
popAndDeath('sex')
#1 is male, 0 is female
```

```python
#Visualize results
df['LPM Prediction'] = res.predict(x)

predictions = df[['sex', 'age_group', 'Race and ethnicity (combined)', 'LPM Prediction', 'death_yn']]

deaths = predictions[predictions['death_yn'] == True]
lives = predictions[predictions['death_yn'] == False]

fig, axs = plt.subplots(1, 2, figsize = (20,8))
fig.suptitle('Fig. 3:\nDistribution of Predictions for the Living and Dead', fontsize= 20)
axs[0].hist(deaths['LPM Prediction'], bins = 10)
axs[0].set_title('Deceased:', fontsize = 18)
axs[0].set_xlabel('Predicted Probability of Death', fontsize= 15)
axs[0].set_ylabel('Frequency', fontsize = 15)

axs[1].hist(lives['LPM Prediction'])
axs[1].set_title('Survived:', fontsize = 18)
axs[1].set_xlabel('Predicted Probability of Death', fontsize = 15)
axs[1].set_ylabel('Frequency', fontsize = 15)

plt.show()
```

```python
dead = pd.DataFrame(df[df['death_yn'] == 1]['LPM Prediction'].describe())
alive = pd.DataFrame(df[df['death_yn'] == 0]['LPM Prediction'].describe())

compare = pd.merge(dead, alive, left_index= True, right_index= True, suffixes = ['_dead', '_alive'])
compare.iloc[1:]
```
