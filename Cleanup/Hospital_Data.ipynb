```python
import pandas as pd
```

```python
df = pd.read_csv('https://healthdata.gov/sites/default/files/reported_hospital_utilization_timeseries_20201213_2141.csv')
df
```

```python
occupancy = df[['date', 'adult_icu_bed_covid_utilization', 'inpatient_beds_utilization']]

def getMonth(date):
    return date[5:7]

occupancy['month'] = occupancy['date'].apply(getMonth)
occupancy = occupancy.groupby('month', dropna = False).mean()[['adult_icu_bed_covid_utilization', 'inpatient_beds_utilization']]
```

```python
# Save Data
df.to_csv('Data/hospital_info.csv', index=False)
occupancy.to_csv('Data/hospital_occupancy.csv', index=False)
```
